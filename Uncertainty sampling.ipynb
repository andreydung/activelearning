```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# create data

mean = [0,0]
cov = [[1,0],[0,100]]
X1 = np.random.multivariate_normal(mean, cov, 200)
Y1 = np.zeros((X1.shape[0]))

mean = [5,0]
cov = [[1,0],[0,50]]
X2 = np.random.multivariate_normal(mean, cov, 200)
Y2 = np.ones((X2.shape[0]))
```

```python
plt.scatter(X1[:,0],X1[:,1])
plt.scatter(X2[:,0],X2[:,1])
plt.show()
```

```python
X = np.vstack([X1, X2])
Y = np.concatenate([Y1, Y2])
```

```python
# random data point for initial classification

index = [1,2,3,4,5,200,201,202,203,204]
```

```python
logistic = linear_model.LogisticRegression()
```

```python
for i in range(10):
    X_labeled = X[index,:]
    Y_labeled = Y[index]
    logistic.fit(X_labeled, Y_labeled)


    Y = logistic.predict(X)
    probabilities = logistic.predict_proba(X)

    utility = np.abs(probabilities[:,0] - 0.5)

    query_index = np.argmin(utility)

    plt.scatter(X[:,0], X[:,1], c=Y)

    plt.scatter(X[query_index, 0], X[query_index, 1], c='r', s=256)

    plt.show()

    print("Query index {}".format(query_index))

    activelabel = input("What is the label of this?")
    print("Active label chosen: {}".format(activelabel))
    
    index.append(query_index)
    Y[query_index] = activelabel

```
